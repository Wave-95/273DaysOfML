```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the function f(x, y)
def f(x, y):
    return x**2 * y + y**2 * x
```

```python
# Generate data for x and y values
x = np.linspace(-10, 10, 21)  # x values ranging from -2 to 2
y = np.linspace(-10, 10, 21)  # y values ranging from -2 to 2
X, Y = np.meshgrid(x, y)     # create a grid of x and y values
```

```python
# Compute the corresponding z values using the function f(x, y)
Z = f(X, Y)
```

```python
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# Plot the function in 3D
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('f(x, y) = x^2 * y + y^2 * x')

# Add rotation and dragging functionality
ax.mouse_init()

# Show the plot
plt.show()
```

```python
x = np.linspace(-10, 10, 21)
y = np.linspace(1, 1, 21)
z = f(x, y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x, y, z)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Surface Plot with Highlighted Slice at y=1')
plt.show()
```
